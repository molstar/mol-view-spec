```typescript
import { createBuilder, MVSJ, molstarNotebook } from "../mod.ts";

const builder = createBuilder();

const structure = builder.download("https://files.wwpdb.org/download/1cbs.cif").parse("mmcif").modelStructure();

structure.component("polymer").representation("cartoon").clip({ref: "clip", type: "plane", point: [22, 13, 0], normal: [0, 0, 1]})

anim = builder.animation()

anim.interpolate(
    kind="scalar",
    target_ref="clip",
    duration_ms=2000,
    property=["point", 2],
    end=55,
    easing="sin-in",
)
```
